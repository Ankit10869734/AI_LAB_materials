```

# Graph representation
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'A': 1, 'D': 1},
    'C': {'A': 3, 'D': 1, 'E': 4},
    'D': {'B': 1, 'C': 1, 'E': 1},
    'E': {'C': 4, 'D': 1}
}
```

```

import heapq
import time

def uniform_cost_search(graph, start, goal):
    start_time = time.time()
    priority_queue = [(0, start, [start])]
    visited = set()

    while priority_queue:
        cost, node, path = heapq.heappop(priority_queue)

        if node in visited:
            continue
        visited.add(node)

        if node == goal:
            return path, cost, time.time() - start_time

        for neighbor, weight in graph[node].items():
            if neighbor not in visited:
                heapq.heappush(priority_queue, (cost + weight, neighbor, path + [neighbor]))

    return None, float('inf'), time.time() - start_time
```

```

from collections import deque

def bfs(graph, start, goal):
    start_time = time.time()
    queue = deque([(start, [start])])
    visited = set([start])

    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path, len(path)-1, time.time() - start_time

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))

    return None, float('inf'), time.time() - start_time


def dfs(graph, start, goal):
    start_time = time.time()
    stack = [(start, [start])]
    visited = set()

    while stack:
        node, path = stack.pop()
        if node in visited:
            continue
        visited.add(node)

        if node == goal:
            return path, len(path)-1, time.time() - start_time

        for neighbor in graph[node]:
            if neighbor not in visited:
                stack.append((neighbor, path + [neighbor]))

    return None, float('inf'), time.time() - start_time
```

```

# Test Case 1
ucs_path, ucs_cost, ucs_time = uniform_cost_search(graph, 'A', 'E')
bfs_path, bfs_cost, bfs_time = bfs(graph, 'A', 'E')
dfs_path, dfs_cost, dfs_time = dfs(graph, 'A', 'E')

print("UCS Path:", ucs_path, "Cost:", ucs_cost, "Time:", ucs_time)
print("BFS Path:", bfs_path, "Cost:", bfs_cost, "Time:", bfs_time)
print("DFS Path:", dfs_path, "Cost:", dfs_cost, "Time:", dfs_time)

# Test Case 2
ucs_path, ucs_cost, _ = uniform_cost_search(graph, 'A', 'C')
print("UCS Path A->C:", ucs_path, "Cost:", ucs_cost)
```

```

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
for node in graph:
    for neighbor, weight in graph[node].items():
        G.add_edge(node, neighbor, weight=weight)

pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, 'weight')

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```
